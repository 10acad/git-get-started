## Challenge_Distribution

In this challenge you consider the real estate data from the city of Ames, Iowa. The details of every real estate transaction in Ames is recorded by the City Assessor's office (https://www.openintro.org/stat/data/ames.csv).

**Based on this data, perform the following tasks**:

- Take a random sample of size 50 from <code>price</code>. Using this sample, what is your best point estimate of the population mean?


- Since you have access to the population, simulate the sampling distribution for the average home price in Ames by taking 5000 samples from the population of size 50 and computing 5000 sample means. Store these means in a vector called <code>sample_means50</code>. Plot the data, then describe the shape of this sampling distribution. Based on this sampling distribution, what would you guess the mean home price of the population to be? Finally, calculate and report the population mean.


- Change your sample size from 50 to 150, then compute the sampling distribution using the same method as above, and store these means in a new vector called <code>sample_means150</code>. Describe the shape of this sampling distribution, and compare it to the sampling distribution for a sample size of 50. Based on this sampling distribution, what would you guess to be the mean sale price of homes in Ames?


- Of the sampling distributions from 2 and 3, which has a smaller spread? If weâ€™re concerned with making estimates that are more often close to the true value, would we prefer a distribution with a large or small spread?


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
%config InlineBackend.figure_format = 'retina'
plt.style.use("seaborn")
```

```python
df = pd.read_csv('ames.csv')
price = df['SalePrice']
```

### Question 1

```python
price.sample(50 , random_state = 42).mean()
```

### Question 2

```python
sample_mean50 = []

for i in range(5000):
    sample_price = np.array(list(price.sample(50).values))
    sample_mean50.append(sample_price.mean())
```

```python
plt.hist(sample_mean50, alpha = 0.5)
plt.show()
```

#### Describe the shape

The mean sample distribution has a large spread with a near like near like normal distribution with the and this is due to the central limit theorem as the number of samples increases the distribution takes a normal form.

The mean is assumed to lie in between the interval of 17500 - 18500 due most of the data point are contained between the interval

```python
print('The best guess of the population mean using the sample mean is : ', np.array(sample_mean50).mean())
print('The population mean is :',price.mean())
```

### Question 3

```python
sample_mean150 = []

for i in range(5000):
    sample_price = np.array(list(price.sample(150).values))
    sample_mean150.append(sample_price.mean())
```

```python
plt.hist(sample_mean150, color = 'g', alpha = 0.5)
plt.show()
```

### Question 3

```python
plt.hist(sample_mean50, alpha = 0.5)
plt.hist(sample_mean150, color = 'g', alpha = 0.5)
plt.show()
```

```python
print('The best guess of the population mean using the sample mean is : ', np.array(sample_mean150).mean())
```

### Question 4

The sample_mean150 have a relatively smaller spread than the sample_mean50. The spread which is shows how far the data points 
are far from the average and if large then the estimate(mean of the sample distribution) is not reliable due to less precision.

A smaller spread shows the data points are closer to the mean reveals a better estimates
