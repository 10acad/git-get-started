#### Question 1. Identify two ways of creating a pandas DataFrame and create a dataframe with ten columns and twenty rows for each way 

```python
import pandas as pd
import numpy as np

# the first dataframe is created using the widely used pandas dataframe creation by 
# specifying the values in a variable and the columns (headers) in a another variable
# the data was generated using the numpy random generator

data = np.random.rand(20,10)
columns = list('ABCDEFGHIJ')


df1 = pd.DataFrame(data , columns = columns) # --------------> first dataframe


# the second dataframe was created using dictionary to hold the data.
# pandas automatically sees the keys as the column header and the values 
# from each key as the row. The rows from this example was stored using the list
# data types and were generated by list operation
# There do say most programmers are very lazy


data2 = {
    
    
    'Name' : ['Gbenga','Femi','Ade','Bose','Bolu','Ayo','David','Esther','Ifeoma','Akeem','Wale','Abeeb','Muraina',
             'Segun','Tosin','Siju','Deji','Blessing','Funmi','Yinka'],
    'Sex' : ['male','female'] * 10 ,
    'Job' : ['Teacher','Banker','Footballer','Trader'] * 5,
    'Job Status' : ['yes'] * 20,
    'Age' : list(np.arange(20,100,4)),
    'weight' : list(np.random.rand(20)),
    'Account' : ['Yes', 'No']  * 10,
    'Transfer' : [0 , 1] * 10,
    'Count' : list(np.arange(20)),
    'Religion' : ['Christian' ,'Muslim'] * 10
}

df2 = pd.DataFrame(data2)
```
