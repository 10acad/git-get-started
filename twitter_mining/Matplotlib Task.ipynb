### BoxPlot

```python
male_ages = data['Q1'][data['Q2'] == 1]
female_ages = data['Q1'][data['Q2'] == 2]
ages = [male_ages , female_ages]
labels = ['Male', 'Female']

plt.boxplot(ages , widths = 0.3, labels = labels , patch_artist = True,
           boxprops = dict(facecolor = 'g')
           )
plt.ylabel('Ages', fontsize = 12)
plt.xlabel('Gender',fontsize = 12)
plt.title("A Boxplot Comparing both Ages of Male and Female Respondents", fontsize = 15)
plt.show();
```

### Barplot

```python
Male = list(data['Q3'][data['Q2'] == 1].value_counts().sort_index())
Female = list(data['Q3'][data['Q2'] == 2].value_counts().sort_index())
ind = np.arange(4) # creates a numpy array from 0 to 3. it will be used on the x-axis
width = 0.4 # will be used to specify the width of the bar
plt.xticks(ind + width / 2, ('Married', 'Divorced', 'Widowed', 'Single'))
plt.bar(ind, Male, width, label="Males")
plt.bar(ind + width, Female, width, label="Females")
plt.title("A Bar Plot comparing the Total numbers of the Relationship Status of Respondents according to their Gender")
plt.xlabel("Marital Status")
plt.ylabel("Total Numbers")
plt.legend()
plt.show()
```

### Histogram

```python
plt.hist(data['Q1'], bins=10 , color = 'c')
plt.title("A Histogram showing the distribution of Ages between 15 and 100")
plt.grid(axis = 'y', alpha = 0.75)
plt.xlabel("Value", fontsize = 12)
plt.ylabel("Frequency", fontsize = 12)
plt.xticks(fontsize = 12)
plt.yticks( fontsize = 12)
plt.show();
```

### Scatter Plot

```python
latitude = data['Latitude']
longitude = data['Longitude']
scale = np.random.rand(200)
scale = 100 * scale ** 2
plt.scatter(latitude, longitude, s=scale)
plt.title("A scatter plot showing the relationship between two lists of randomly generated integers between 150 and 275")
plt.show()
```
