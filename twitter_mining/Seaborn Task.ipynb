Import the pandas_tasks.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()

data = pd.read_csv("pandas_tasks.csv")
```

#### Question 1. Make a scatter plot of the Longitude and Latitude columns

```python
sns.relplot(x="Longitude", y="Latitude", height= 8, aspect=1, data=data)
```

#### Question 2. Make a scatter plot of the Longitude and Latitude columns. This time, plot with 'hue' = Q2 and compare it over Q3

```python
sns.relplot(x="Longitude", y="Latitude", height= 8, aspect=1, hue="Q2",data=data)
```

#### Question 3. Make a boxplot of Longitude and Q2

```python
sns.boxplot(x='Q2', y='Longitude', data=data)
```
